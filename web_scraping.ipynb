```python
years = range(2001, 2024)
#range of years to collect MVP data
```

```python
awards_url = "https://www.pro-football-reference.com/awards/awards_{}.htm"
```

```python
import requests

for year in years:
    url = awards_url.format(year)
    data = requests.get(url)
    
    with open("mvp/{}.html".format(year), "w+") as file:
        file.write(data.text)
        #scrapes data year by year
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
dfs = []
for year in years:
    with open("mvp/{}.html".format(year)) as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="over_header").decompose()
    mvp_table = soup.find(id="voting_apmvp")
    mvp = pd.read_html(str(mvp_table))[0]
    mvp["Year"] = year
    
    dfs.append(mvp)
```

```python
mvps = pd.concat(dfs)
```

```python
mvps
```

```python
mvps.to_csv("mvps.csv")
```

```python
driver = webdriver.Chrome(executable_path="/Users/rishabjadhav/chromedriver")
```

```python
passing_stats_url = "https://www.pro-football-reference.com/years/{}/passing.htm"
```

```python
import time

url = passing_stats_url.format(2023)

driver.get(url)
driver.execute_script("window.scrollTo(1,10000)")
time.sleep(2)

html = driver.page_source

with open("passing/{}.html".format(2023), "w+") as f:
    f.write(html)
```

```python
dfs = []

for year in years:
    with open("passing/{}.html".format(year)) as f:
        page = f.read()
        
    soup = BeautifulSoup(page, "html.parser")
    soup.find("tr", class_="thead").decompose()
    passing_table = soup.find(id="passing")
    passing = pd.read_html(str(passing_table))[0]
    passing["Year"] = year
    dfs.append(passing)
```

```python
passing = pd.concat(dfs)
```

```python
rushing_stats_url = "https://www.pro-football-reference.com/years/{}/rushing.htm"
```

```python
import time

url = rushing_stats_url.format(2023)

driver.get(url)
driver.execute_script("window.scrollTo(1,10000)")
time.sleep(2)

html = driver.page_source

with open("rushing/{}.html".format(2023), "w+") as f:
    f.write(html)
```

```python
dfs = []

for year in years:
    with open("rushing/{}.html".format(year)) as f:
        page = f.read()
        
    soup = BeautifulSoup(page, "html.parser")
    soup.find("tr", class_="thead").decompose()
    soup.find("tr", class_="over_header").decompose()
    
    rushing_table = soup.find(id="rushing")
    rushing = pd.read_html(str(rushing_table))[0]
    rushing["Year"] = year
    dfs.append(rushing)
```

```python
rushing = pd.concat(dfs)
rushing.to_csv("rushing.csv")
```
