```python
years = list(range(2013, 2022))
```

```python
awards_url = "https://www.pro-football-reference.com/awards/awards_{}.htm"
```

```python
import requests

for year in years:
    url = awards_url.format(year)
    data = requests.get(url)
    
    with open("mvp/{}.html".format(year), "w+") as file:
        file.write(data.text)
```

```python
dfs = []
for year in years:
    with open("mvp/{}.html".format(year)) as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="over_header").decompose()
    mvp_table = soup.find(id="voting_apmvp")
    mvp = pd.read_html(str(mvp_table))[0]
    mvp["Year"] = year
    
    dfs.append(mvp)
```

```python
mvps = pd.concat(dfs)
```

```python
mvps
```

```python
mvps.to_csv("mvps.csv")
```

```python
passing_stats_url = "https://www.pro-football-reference.com/years/{}/passing.htm"
```

```python
for year in years:
    url = passing_stats_url.format(year)
    data = requests.get(url)
    
    with open("passing/{}.html".format(year), "w+") as file:
        file.write(data.text)
```

```python
url = passing_stats_url.format(2022)
data = requests.get(url)
    
with open("passing/2022.html".format(2022), "w+") as file:
    file.write(data.text)
```
