```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
mvps = mvps[["Player", "Year", "Votes", "Share"]]
```

```python
mvps
```

```python
passing = pd.read_csv("passing.csv")
```

```python
passing
```

```python
del passing["Unnamed: 0"]
del passing["Rk"]
```

```python
passing.head()
```

```python
passing["Player"] = passing["Player"].str.replace("*", "", regex = False)
```

```python
passing.head()
```

```python
passing["Player"] = passing["Player"].str.replace("+", "", regex = False)
```

```python
passing.head(50)
```

```python
combined = passing.merge(mvps, how = "outer", on = ["Player", "Year"])
```

```python
combined.head(10)
```

```python
combined[combined["Votes"] > 0]
```

```python
combined[["Votes", "Share"]] = combined[["Votes", "Share"]].fillna(0)
```

```python
combined[combined["Votes"] > 0]
```

```python
combined[combined["Pos"] != "QB"].tail(20)
```
