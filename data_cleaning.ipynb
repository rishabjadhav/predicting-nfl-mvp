```python
#to combine these we need to differentiate passing and rushing yards, touchdowns, etc
passing = passing.rename(columns={"Att": "pAtt",})
passing = passing.rename(columns={"Yds": "pYds",})
passing = passing.rename(columns={"TD": "pTD",})
passing = passing.rename(columns={"1D": "p1D",})
passing = passing.rename(columns={"Lng": "pLng",})
passing = passing.rename(columns={"Y/A": "pY/A",})
passing = passing.rename(columns={"Y/G": "pY/g",})
```

```python
#same thing for rushing
rushing = rushing.rename(columns={"Att": "rAtt",})
rushing = rushing.rename(columns={"Yds": "rYds",})
rushing = rushing.rename(columns={"TD": "rTD",})
rushing = rushing.rename(columns={"1D": "r1D",})
rushing = rushing.rename(columns={"Lng": "rLng",})
rushing = rushing.rename(columns={"Y/A": "rY/A",})
rushing = rushing.rename(columns={"Y/G": "rY/g",})
```

```python
rushing = pd.read_csv("rushing_cleaned.csv")
passing = pd.read_csv("passing_cleaned.csv")
```

```python
stats = passing.merge(rushing, how = "outer", on = ["Player", "Year"])
```

```python
stats
```

```python
#many rushing players have no passing value data, and have NaNs, they can be dropped
```

```python
stats = stats.dropna()
```

```python
stats
```

```python
pd.isnull(stats).sum()
```

```python
stats.columns
```

```python
del stats["Unnamed: 0_x"]
del stats["Unnamed: 0_y"]
del stats["Age_y"]
del stats["G_y"]
del stats["GS_y"]
```

```python
stats = stats.rename(columns={"Age_x": "Age",})
stats = stats.rename(columns={"G_x": "G",})
stats = stats.rename(columns={"GS_x": "GS",})
```

```python
stats.columns
```

```python
#now to combine with the mvp data to the stats to complete our composite training and testing data!
```

```python
stats.dtypes
```

```python
findata = stats.merge(mvps, how = "outer", on = ["Player", "Year"])
```

```python
findata
```

```python
pd.isnull(findata).sum()
```

```python
findata["Share"] = findata["Share"].fillna(0)
```

```python
findata
```

```python
#the remaining 21 nulls are all either defensive players or do not have any passing or rushing data
#therefore they can be dropped
findata = findata.dropna()
```
